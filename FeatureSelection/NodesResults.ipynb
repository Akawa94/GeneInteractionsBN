```python
import pandas as pd
import numpy as np
import bz2
import pickle
from ipynb.fs.defs.GraphReader import GraphReader
import seaborn as sns
```

```python
def read_results(selected_node):
    save_route = "/home/a20114261/sdelrio/GeneInteractions/GeneInteractionsBN_Datasets/P1_results" 
    with bz2.open(save_route+"/comparisonLT_Alarm5_s500_v5_node"+selected_node+"_.bz2") as f:
        r = pickle.load(f)
    gr = GraphReader("/home/a20114261/sdelrio/alarm_datasets/Alarm5/Alarm5_graph.txt")
    mm_keys = ["MMPCOPT","MMRWPC"]
    r_dict = {}
    r_dict["node"] = selected_node
    node_index = int(selected_node)-1
    for key in mm_keys:
        r_dict[key+"_Sum"] = sum([gr.getDistance(node_index,[y.split("@")[0] for y in x[key]["nodes"]])["distance-pc"] for x in r["results"]])/10
        r_dict[key+"_StdDev"] = sum([(gr.getDistance(node_index,[y.split("@")[0] for y in x[key]["nodes"]])["distance-pc"]-r_dict[key+"_Sum"])**2 for x in r["results"]])/10
    return r_dict
```

```python
results = []
for i in range(0,185):
    try:
        results.append(read_results(str(i+1)))
    except:
        print("failed try on Node"+str(i+1))
        
        
```

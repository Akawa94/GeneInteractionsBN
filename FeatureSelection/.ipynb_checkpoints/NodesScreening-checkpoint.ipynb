```python
# will be completed after downloading remaining datasets
datasets_dir = {
    'alarm':{
        'data':"/home/a20114261/sdelrio/alarm_datasets/Alarm5",
    },
    'child':{},
    'hildfinder': {},
}
```

```python
for key in datasets_dir:
    if (len(datasets_dir[key])==0):
        continue
    directory = datasets_dir[key]['data']
    graph_path = datasets_dir[key]['data']+'/'+datasets_dir[key]['data'].split('/')[len(datasets_dir[key]['data'].split('/'))-1]+'_graph.txt'
    print(graph_path)
    gr = GraphReader(graph_path)
    lower_datasets = {}
    for file in os.listdir(directory):
        if (file.split('_')[1]=='s500'):
            print(file)
            X = pd.read_csv(directory+'/'+file,delimiter='  ',header=None, engine='python')
            g_heads=[]
            for e in X:
                g_heads.append('Node'+str(e+1)+'_'+str(len(set(X[e]))))
            X.columns = g_heads
            x_dict={}
            x_dict['dataset'] = X
            lower_datasets[file] = x_dict
```

```python
for key in lower_datasets:
    lower_datasets[key]['balance_arr'] = {}
    X = lower_datasets[key]['dataset']
    for s_key in X:
        # retrieve existing categories
        categories = set(X[s_key])
        print(s_key)
        print(categories)
        balance_arr = []
        for cat in categories:
            cat_count = list(X[s_key]).count(cat)
            balance_arr.append({
                'category':cat,
                'percentage':cat_count*len(categories)/500
            })
        lower_datasets[key]['balance_arr'][s_key] = balance_arr
```

```python
preliminary_node_dict = {}
for key in lower_datasets:
    for s_key in lower_datasets[key]['dataset']:
        if (s_key.split('_')[0] not in list(preliminary_node_dict.keys())):
            preliminary_node_dict[s_key.split('_')[0]]={}
        for c_dict in lower_datasets[key]['balance_arr'][s_key]:
            print(c_dict)
            if (c_dict['category'] not in preliminary_node_dict[s_key.split('_')[0]]):
                preliminary_node_dict[s_key.split('_')[0]][c_dict['category']] = 0
            preliminary_node_dict[s_key.split('_')[0]][c_dict['category']] += c_dict['percentage']
```

```python
final_node_list = []
for key in preliminary_node_dict:
    below_threshold = False
    for s_key in preliminary_node_dict[key]:
        print(preliminary_node_dict[key][s_key])
        if preliminary_node_dict[key][s_key]< 0.05*2*10:
            below_threshold = True
            break
    if (below_threshold == True):
        final_node_list.append(key)
            
```

```python
from ipynb.fs.defs.MMRWPC_v2 import MMRWPC
from ipynb.fs.defs.MMPC_opt import MMPC_opt
from sklearn.feature_selection import mutual_info_classif as mi_clasf
from scipy.stats import ttest_ind
from scipy.stats import ttest_rel
import time
import itertools
```

```python
def process_MMRWPC(MMPC_RW_model, node):
    result = MMPC_RW_model.evaluate(node)
    node_name = "Node"+str(node+1)    
    
    return {
        'name':node_name,
        'nodes':[x['name'] for x in itertools.chain.from_iterable(result['CPC'])],
        'time':result['time']
    }
```

```python
def process_MMPC_opt(MMPC_opt_model, node):
    result = MMPC_opt_model.evaluate(node)
    node_name = "Node"+str(node+1)
    return {
        'name':node_name,
        'nodes':[x['name'] for x in result['CPC']],
        'time':result['time']
    }
```

```python
def process_batch_params(params):
    i, MMPCOPT_model, MMRWPC_model, node_index = params
    print("Processing try n "+str(i))
    sub_results = {}
    MMPCOPT_model.shuffle_rows()
    MMRWPC_model.recieveX(MMPCOPT_model.deliverX())
    sub_results['MMRWPC'] = process_MMRWPC(MMRWPC_model, node_index)
    sub_results['MMPCOPT'] = process_MMPC_opt(MMPCOPT_model, node_index)
    print(sub_results)
    return sub_results
```

```python
def process_batch(batch_str, node_index):
    X = pd.read_csv(batch_str,delimiter='  ',header=None, engine='python')
    g_heads=[]
    for e in X:
        g_heads.append('Node'+str(e)+'_'+str(len(set(X[e]))))
    X.columns = g_heads
    results = []
    MMPCOPT_model = MMPC_opt(0.05)
    MMPCOPT_model.fit(X)
    MMRWPC_model = MMRWPC(0.05,0.0,1)
    MMRWPC_model.fit(X)
    pool = multiprocessing.Pool(10)
    params = []
    for i in range(0,10):
        sub_params = [i,MMPCOPT_model,MMRWPC_model, node_index]
        params.append(sub_params)
    results = pool.map(process_batch_params,params)
    pool.close()
    pool.join()
    return results
```

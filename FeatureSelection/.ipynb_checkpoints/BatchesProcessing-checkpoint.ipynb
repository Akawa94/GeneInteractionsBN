```python
import pandas as pd
import numpy as np
import os
import itertools
from ipynb.fs.defs.MMRWPC_v2 import MMRWPC
from ipynb.fs.defs.MMPC_opt import MMPC_opt
from ipynb.fs.defs.CrossAdaptiveGA import *
from sklearn.feature_selection import mutual_info_classif as mi_clasf
from scipy.stats import ttest_ind
from scipy.stats import ttest_rel
import pickle
import multiprocessing
import time
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVR
import bz2 # for pickling and unplicking large files
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def process_MMRWPC(X):
#     global x_train
#     global x_test
    t_start = time.time()
    MMPC_RW_model = MMRWPC(0.05,0.0,1)
    MMPC_RW_model.fit(X)
    result = MMPC_RW_model.evaluate(0)
    nodes = {}
    nodes['Node1@2']={
        'nodes':[x['name'] for x in itertools.chain.from_iterable(result['CPC'])],
        'time':result['time']
                   }
    t_stop = time.time()
    
    return {
        'nodes':nodes,
        'time':t_start-t_stop,
    }
```

```python
def process_MMPC_opt(batch_str):
    X = pd.read_csv(batch_str,delimiter=',', engine='python')
    X['Pvalue'] = list(map(lambda x: 1 if x<=0.01 else 0,list(X['Pvalue'])))
    del X['CauseGene']
    del X['EffectGene']
    del X['Replicate']
    del X['Treatment']
    t_start = time.time()
    MMPC_opt_model = MMPC_opt(0.05)
    MMPC_opt_model.fit(X)
    result = MMPC_opt_model.evaluate(0)
    nodes = {}
    nodes['Node1@2']={
        'nodes':[x['name'] for x in result['CPC']],
        'time':result['time']
                   }
    t_stop = time.time()
    
    return {
        'nodes':nodes,
        'time':t_start-t_stop,
    }
```

```python
def process_MI(X):
    t_start = time.time()
    mi_score = mi_clasf(X, X[list(X.keys())[0]],discrete_features=True)
    mi_results = list(zip(list(map(lambda x:'Node'+str(x+1),np.arange(1,541).tolist())),mi_score[1:]))
    t_stop = time.time()
    return {
        'nodes':mi_results,
        'time':t_stop-t_start,
    }
```

```python
def process_TTEST_ind(X):
    t_start = time.time()
    ttest_results = []
    x_keys = list(X.keys())
    x_pvalue_key = x_keys[0]
    x_keys.pop(0)
    for i in range(0,len(x_keys)):
        ttest_results.append(['Node'+str(i+2),ttest_ind(X[x_pvalue_key],X[x_keys[i]])[1]])
    t_stop = time.time()
    return {
        'nodes':ttest_results,
        'time':t_stop-t_start,
    }
```

```python
def process_TTEST_rel(X):
    t_start = time.time()
    ttest_results = []
    x_keys = list(X.keys())
    x_pvalue_key = x_keys[0]
    x_keys.pop(0)
    for i in range(0,len(x_keys)):
        ttest_results.append(['Node'+str(i+2),ttest_rel(X[x_pvalue_key],X[x_keys[i]])[1]])
    t_stop = time.time()
    return {
        'nodes':ttest_results,
        'time':t_stop-t_start,
    }
```

```python
def process_batch(batch_str):
    X = pd.read_csv(batch_str,delimiter=',', engine='python')
    X['Pvalue'] = list(map(lambda x: 1 if x<=0.01 else 0,list(X['Pvalue'])))
    del X['CauseGene']
    del X['EffectGene']
    del X['Replicate']
    del X['Treatment']
    
    
    results = {}
    results['MI'] = process_MI(X)
    results['TTEST_ind'] = process_TTEST_ind(X)
    results['TTEST_rel'] = process_TTEST_rel(X)
    results['MMRWPC'] = process_MMRWPC(X)
    return results
    
```

```python
def making_x_train(directory):
    batch_arr = []
    for batch_str in os.listdir(directory):
        if (len(batch_str.split('batch'))==1):
            continue
        X = pd.read_csv(directory+'/'+batch_str,delimiter=',', engine='python')
        X['Pvalue'] = list(map(lambda x: 1 if x<=0.01 else 0,list(X['Pvalue'])))
        del X['CauseGene']
        del X['EffectGene']
        del X['Replicate']
        del X['Treatment']
        batch_arr.append(X)
    X_train = pd.concat(batch_arr)
    print(X_train.shape)
    return X_train
```

```python
def model_memory_scorer(filename, model_name,train_data, test_data, target_node, nodes_lists):
    results = {}
    train_data.columns = test_data.columns
    for key in nodes_lists:
        X_train = train_data
        Y_train = train_data[target_node]
        if (model_name == 'NB'):
            clf = GaussianNB()
        elif (model_name == 'LR'):
            clf = LogisticRegression()
        elif (model_name == 'DTC'):
            clf = DecisionTreeClassifier()
        elif (model_name == 'SVR'):
            clf = SVR()
        clf.fit(X_train[nodes_lists[key]],Y_train)
        precisions = []
        memory = []
        values_counter=set(train_data[target_node])
        for val in values_counter:
            if (len(test_data.query(target_node+'== '+str(val)))==0): 
                precisions.append(1)
                continue
            Y_pred_c = clf.predict(test_data.query(target_node+'== '+str(val))[nodes_lists[key]])
            precisions.append(Y_pred_c.tolist().count(val)/len(Y_pred_c))
            memory.append({'value':val, 'pred': Y_pred_c})
        results[key] = {'g-mean':reduce(operator.mul,precisions,1)**(1/len(precisions)),
                       'memory':memory,
                       'filename':filename,
                        'nodes': nodes_lists[key]
                       }
    return results    
```

```python
def making_x_single(directory, batch_str):
    X = pd.read_csv(directory+'/'+batch_str,delimiter=',', engine='python')
    X['Pvalue'] = list(map(lambda x: 1 if x<=0.01 else 0,list(X['Pvalue'])))
    del X['CauseGene']
    del X['EffectGene']
    del X['Replicate']
    del X['Treatment']
    return X
```

```python
def get_single_tops(batch):
    results_dict = {}
    key_list = list(batch.keys())
    for key in key_list:
        results = np.zeros((541,)).tolist()
        print(key)
        if('PC' not in key ):
            for node in batch[key]['nodes']:
                index = int(node[0].split('Node')[1])-1
                value = node[1]                    
                results[index] += value
            unsorted_list = list(zip(['Node'+str(x+1) for x in np.arange(0,541).tolist()],results))
            sorted_list = sorted(unsorted_list, key = lambda x : x[1], reverse =  True)
            results_dict[key] = {
                'top2':sorted_list[:1],
                'top3':sorted_list[:3],
                'top5':sorted_list[:5],
                'top10':sorted_list[:10],
                'top15':sorted_list[:15]
            }
        else:
            for node in batch[key]['nodes']['Node1@2']['nodes']:                    
                index = int(node.split('Node')[1].split('@')[0])-1
                value = 1                    
                results[index] += value
            unsorted_list = list(zip(['Node'+str(x+1) for x in np.arange(0,541).tolist()],results))
            sorted_list = [x for x in unsorted_list if x[1] > 0]
            results_dict[key] = {
                'pc':sorted_list,
            }
    return results_dict
```

```python
for directory in os.listdir(main_directory):
    if '.pickle' in directory or 'censored' not in directory or '.csv' in directory:
        continue
    results_df = pd.DataFrame(columns=["FeatureSelector","Classifier","Top","GMean"])
    file_pre = main_directory+results_save_folder+directory
    s_results = {}
    for i in range(0,100):
        print(file_pre+'_sr_p'+str(i)+'.bz2')
        with bz2.open(file_pre+'_sr_p'+str(i)+'.bz2','rb',9) as f:
            s_data = pickle.load(f)
        print(i)
        for key in (list(s_data.keys())):
            print(key)
            if (key in list(s_results.keys())):
                for s_key in list(s_data[key].keys()):
                    for t_key in list(s_data[key][s_key].keys()):
                        for e_arr in list(s_data[key][s_key][t_key]['memory']):
                            s_results[key][s_key][t_key][e_arr['value']]['pred'] = \
                            [sum(x) for x in zip(s_results[key][s_key][t_key][e_arr['value']]['pred'], 
                                                 [x*s_data[key][s_key][t_key]['g-mean'] for x in e_arr['pred']])]
                            s_results[key][s_key][t_key]['weight'] += s_data[key][s_key][t_key]['g-mean']
                            #print(len(s_results[key][s_key][t_key][e_arr['value']]['pred']))
            else:
                s_results[key] = {}
                for s_key in list(s_data[key].keys()):
                    s_results[key][s_key] = {}
                    for t_key in list(s_data[key][s_key].keys()):
                        s_results[key][s_key][t_key] = {}
                        s_results[key][s_key][t_key]['weight'] = s_data[key][s_key][t_key]['g-mean']
                        for e_arr in s_data[key][s_key][t_key]['memory']:
                            s_results[key][s_key][t_key][e_arr['value']] = {}
                            s_results[key][s_key][t_key][e_arr['value']]['pred'] = \
                            [x*s_data[key][s_key][t_key]['g-mean'] for x in e_arr['pred']]    
    for key in (list(s_results.keys())):
        for s_key  in (list(s_results[key].keys())):
            for t_key in (list(s_results[key][s_key].keys())):
                acc = 1
                root_n = 0
                for v_key in (list(s_results[key][s_key][t_key])):
                    s_results[key][s_key][t_key][v_key]['result'] = \
                    list(map(lambda x: 1 if x/s_results[key][s_key][t_key]['weight']>=0.5 else 0,s_results[key][s_key][t_key][v_key]['pred'])).count(v_key)\
                    /len(s_results[key][s_key][t_key][v_key]['pred'])
                    acc *= s_results[key][s_key][t_key][v_key]['result']
                    root_n += 1
                s_results[key][s_key][t_key]['g-mean'] = acc**(1/root_n)
                results_df = results_df.append({
                    "Classifier":s_key,
                    "FeatureSelector":key,
                    "Top":t_key,
                    "GMean": s_results[key][s_key][t_key]['g-mean']
                }, ignore_index=True)
    results_df.to_csv(main_directory+directory+'_ensemble_results_v2.csv',index=False)
```

```python
def get_tops(key_list, f_selection_data):
    results_dict = {}
    
    for key in key_list:
        results = np.zeros((541,)).tolist()
        print(key)
        if('PC' not in key ):
            for sub_f in f_selection_data:
                for node in sub_f[key]['nodes']:
                    index = int(node[0].split('Node')[1])-1
                    value = node[1]                    
                    results[index] += value
            unsorted_list = list(zip(['Node'+str(x+1) for x in np.arange(0,541).tolist()],results))
            sorted_list = sorted(unsorted_list, key = lambda x : x[1], reverse =  True)            
        else:
            for sub_f in f_selection_data:
                for node in sub_f[key]['nodes']['Node1@2']['nodes']:                    
                    index = int(node.split('Node')[1].split('@')[0])-1
                    value = 1                    
                    results[index] += value
            unsorted_list = list(zip(['Node'+str(x+1) for x in np.arange(0,541).tolist()],results))
            sorted_list = sorted(unsorted_list, key = lambda x : x[1], reverse =  True)
        results_dict[key] = {
            'top2':sorted_list[:1],
            'top3':sorted_list[:3],
            'top5':sorted_list[:5],
            'top10':sorted_list[:10],
            'top15':sorted_list[:15]
        }
    return results_dict
```

```python
def get_node_set(key_list, f_selection_data):
    results_dict = {}
    
    for key in key_list:
        node_set = []
        if('PC' not in key ):
            for sub_f in f_selection_data:
                for node in sub_f[key]['nodes']:
                    index = int(node[0].split('Node')[1])-1
                    value = node[1]                    
                    results[index] += value
            unsorted_list = list(zip(['Node'+str(x+1) for x in np.arange(0,541).tolist()],results))
            sorted_list = sorted(unsorted_list, key = lambda x : x[1], reverse =  True)            
        else:
            for sub_f in f_selection_data:
                for node in sub_f[key]['nodes']['Node1@2']['nodes']:                    
                    index = int(node.split('Node')[1].split('@')[0])-1
                    value = 1                    
                    results[index] += value
            unsorted_list = list(zip(['Node'+str(x+1) for x in np.arange(0,541).tolist()],results))
            sorted_list = sorted(unsorted_list, key = lambda x : x[1], reverse =  True)
        results_dict[key] = {
            'top2':sorted_list[:1],
            'top3':sorted_list[:3],
            'top5':sorted_list[:5],
            'top10':sorted_list[:10],
            'top15':sorted_list[:15]
        }
    return results_dict
```

```python
def model_scorer(model_name,train_data, test_data, target_node, nodes_lists):
    
    results = {}
    for key in nodes_lists:
        X_train = train_data
        Y_train = train_data[target_node]
        if (model_name == 'NB'):
            clf = GaussianNB()
        elif (model_name == 'LR'):
            clf = LogisticRegression()
        elif (model_name == 'DTC'):
            clf = DecisionTreeClassifier()
        clf.fit(X_train[nodes_lists[key]],Y_train)

        precisions = []
        values_counter=set(train_data[target_node])
        for val in values_counter:
            if (len(test_data.query(target_node+'== '+str(val)))==0): 
                precisions.append(1)
                continue
            Y_pred_c = clf.predict(test_data.query(target_node+'== '+str(val))[nodes_lists[key]])
            precisions.append(Y_pred_c.tolist().count(val)/len(Y_pred_c))
        results[key] = reduce(operator.mul,precisions,1)**(1/len(precisions))
    return results

    
```

```python
def GetCondValues(z, condVars):
    if len(condVars)>1:
        maxz = reduce(operator.mul,[int(x['name'].split('@')[1]) for x in condVars],1)
        d = np.zeros(len(condVars))
        
        div_dim = maxz/int(condVars[0]['name'].split('@')[1])
        num2div = z
        for i in range(0,len(condVars)-1):
            d[i] = floor(num2div/div_dim)
            num2div=num2div%div_dim
            div_dim = div_dim/int(condVars[i+1]['name'].split('@')[1])
        d[i+1] = num2div
    else:
        d = [z]
    return d
```

```python
def dep_dict_retriever(x_node,v_look):
    global dep_dict
    
    # making hash key
    x_node_key = x_node['name']+'_'+str(v_look)
    if x_node_key in dep_dict:
        return dep_dict[x_node_key]
    else:
        returnable = np.array(list(map(lambda x: 1 if x==v_look else 0, x_node['data'])))
        dep_dict[x_node_key] = returnable
        return returnable
```

```python
class MultNode:
    def __init__(self,parent,name,result):
        self.parent = parent
        self.name = name
        self.result = result
        self.mult_dict = {}
```

```python
def mult_dict_retrivier(node_list,values_list):
    # most  efficient approach?
    global mult_dict
    list_size = len(node_list)
    list_up_counter = 0
    vector_result=[]
    curr_mult_node = MultNode('none','none',-1)
    
    while list_up_counter<list_size:
        # basic case
        mult_dict_key = node_list[list_up_counter]['name']+'_'+str(values_list[list_up_counter])
        if mult_dict_key in mult_dict:
            # traverse through it
            curr_mult_node = mult_dict[mult_dict_key]
            while (list_up_counter<list_size) and not_found_flag==-1:
                    
                    vector_result = curr_mult_node.result
                    next_key = node_list[list_up_counter]['name']+'_'+str(values_list[list_up_counter])
                    if next_key in curr_mult_node.mult_dict:
                        curr_mult_node = curr_mult_node.mult_dict[next_key]
                        #list_up_counter += 1
                    else:
                        # 2 ways out:
                        # no node in initial dict
                        if next_key not in mult_dict:
                            n_result = np.array(list(map(lambda x: 1 if x==values_list[list_up_counter] else 0, node_list[list_up_counter]['data'])))
                            mult_dict[next_key] = MultNode('none',next_key,n_result)
                            curr_mult_node.dict[next_key] = MultNode(curr_mult_node.name,next_key,curr_mult_node.result*n_result)
                            curr_mult_node = curr_mult_node.dict[next_key]
                            
                        # node in initial dict
                        else:
                            curr_mult_node.dict[next_key] = MultNode(curr_mult_node.name,next_key,mult_dict[next_key].result*n_result)
                            curr_mult_node = curr_mult_node.dict[next_key]
                    list_up_counter += 1
            
        else:
            # calculate value and store it
            if (curr_mult_node.name == 'none'): # first found result
                vector_result = np.array(list(map(lambda x: 1 if x==values_list[list_up_counter] else 0, node_list[list_up_counter]['data'])))
                mult_dict[mult_dict_key] = MultNode('none',mult_dict_key,vector_result)
                curr_mult_node = mult_dict[mult_dict_key]
            else: # not first value encountered
                n_result = np.array(list(map(lambda x: 1 if x==values_list[list_up_counter] else 0, node_list[list_up_counter]['data'])))
                mult_dict[mult_dict_key] = MultNode('none',mult_dict_key,n_result)
                curr_mult_node.dict[mult_dict_key] = MultNode(curr_mult_node.name,mult_dict_key,curr_mult_node.result*n_result)
                curr_mult_node = curr_mult_node.dict[mult_dict_key]
            list_up_counter+=1
    return curr_mult_node.result
```

```python
def Dep_v2(TargetNode, Xi, CondVars, X, alpha):
    # Dep_v2
    #print('in Dep')
    global data_c_size
    global dep_dict
    CondVars = sorted(CondVars,key=operator.itemgetter('name'))
    
    vvTarget = []
    for i in range(int(TargetNode['name'].split('@')[1])):
        vvTarget.append(i)
    
    vvXi = []
    for i in range(int(Xi['name'].split('@')[1])):
        vvXi.append(i)
    
    szCondVars = reduce(operator.mul,[int(x['name'].split('@')[1]) for x in CondVars],1)
    
    
    if data_c_size <= (5 * len(vvTarget)*len(vvXi)*(szCondVars)):
        return 1
    
    if (len(CondVars)==0):
        S = np.zeros((len(vvTarget),len(vvXi)))
        for i in range(0,len(vvTarget)):
            for j in range(0,len(vvXi)):
#                 op1 = dep_dict_retriever(TargetNode,vvTarget[i])
#                 op2 = dep_dict_retriever(Xi,vvXi[j])
                S[i][j]=np.sum(mult_dict_retriever([TargetNode,Xi],[vvTarget[i],vvXi[j]]))
        G2=0
        N = np.sum(S)
#         print(S)
#         print(N)
        Si =np.sum(S,axis=1)
        Sj =np.sum(S,axis=0)
        Df = ((len(vvTarget)-1)*(len(vvXi)-1))
#         print(S[np.where(S>0)])
        Dedf = len((S[np.where(S>0)]))\
                      - len(Si[np.where(Si>0)])\
                      - len(Sj[np.where(Sj>0)]) +1
#         print("Dedf before if: ",Dedf)
        if (Dedf<1):
            Dedf=1            
            #return 0
        #print(S)
        for i in range(0,len(vvTarget)):
            for j in range(0,len(vvXi)):
                #print(i)
                #print(j)
                if (S[i][j]>0):
                    #print(Si[i])
                    #print(Sj[j])
                    G2 = G2 + S[i][j]*np.log((S[i][j])*N/(Si[i]*Sj[j]))
                    #print(G2)
        
    else: # test conditional dependency
        S = np.zeros((len(vvTarget),len(vvXi),szCondVars))
        for i in range(0,len(vvTarget)):
            for j in range(0,len(vvXi)):
                for k in range(0,szCondVars):
                    condValue = GetCondValues(k,CondVars)
                    #print(condValue)
                    cond_vars_arr = []
                    for l in range(0,len(CondVars)):
                        cond_vars_arr.append(CondVars[l])
                    
                    op1_op2_flagDataCondVars = mult_dict_retriever(\
                                    cond_vars_arr+[TargetNode,Xi],\
                                    CondVars+[vvTarget[i],vvXi[j]])
                    #print("Omg we got out!")
                    
                        
#                         op_x = dep_dict_retriever(X_l,condValue[l])

#                         flagDataCondVars = flagDataCondVars*op_x
                        
#                     op1 = dep_dict_retriever(TargetNode,vvTarget[i])
                    
#                     op2 = dep_dict_retriever(Xi,vvXi[j])
                    
                    #prod_op1_op2 = mult_dict_retriever([TargetNode,Xi],[vvTarget[i],vvXi[j]])
                    S[i][j][k]=np.sum(op1_op2_flagDataCondVars)
                    #S[i][j][k]=np.sum(op1*op2*flagDataCondVars)
        G2 = 0
        
        Sjk = np.sum(S,axis=0)
        Sik = np.sum(S,axis=1)
        Sk = np.sum(Sjk,axis=0)
        
        #Sjk = sum(S,1);
        #Sik = sum(S,2);
        #Sk = sum(Sjk,2); 
        Dedf = len(S[np.where(S>0)]) -\
                len(Sik[np.where(Sik>0)]) -\
                len(Sjk[np.where(Sjk>0)]) +\
                len(Sk[np.where(Sk>0)])
        if Dedf<1:
            Dedf=1
            #print('No, here is the trouble')
            #return 0
        
        # compute G2 statistic
        #print(S.shape)
        #print(Sjk.shape)
        #print(Sik.shape)
        #print(Sk.shape)
        for i in range(0,len(vvTarget)):
            for j in range(0,len(vvXi)):
                for k in range(0,szCondVars):
                    if (S[i][j][k]>0):
                        #print(i,j,k)
                        #test_op =  Sk[k]
                        #test_op = Sik[i][k]
                        #test_op = Sjk[j][k]
                        G2 += (S[i][j][k] * np.log(\
                                    (S[i][j][k] * Sk[k])/\
                                    (Sik[i][k] * Sjk[j][k])))
        G2 = 2*G2
    
    
    
    assoc = (alpha - (1 - chi2.cdf(G2,Dedf)))/alpha
    #print(G2)
    #print(Dedf)
    #print(assoc)
    if assoc<0:
        assoc=0
#     print("Target Node: ",TargetNode['name'])
#     print("Node XI: ", Xi['name'])
#     print("\tG2:",G2)
#     print("\tDedf: ",Dedf)
#     print("\tAssoc: ",assoc)
#     print()
#     print()
    return assoc
```

```python
def Dep(TargetNode, Xi, CondVars, X, alpha):
    #Dep_v4_gpu
    #print('in Dep')
    global data_c_size
    global dep_dict
    
    vvTarget = []
    for i in range(int(TargetNode['name'].split('@')[1])):
        vvTarget.append(i)
    
    vvXi = []
    for i in range(int(Xi['name'].split('@')[1])):
        vvXi.append(i)
    
    szCondVars = reduce(operator.mul,[int(x['name'].split('@')[1]) for x in CondVars],1)
    
    
    if data_c_size <= (5 * len(vvTarget)*len(vvXi)*(szCondVars)):
        return 1
    
    if (len(CondVars)==0):
        S = np.zeros((len(vvTarget),len(vvXi)))
        for i in range(0,len(vvTarget)):
            for j in range(0,len(vvXi)):
                op1 = dep_dict_retriever(TargetNode,vvTarget[i])
                op2 = dep_dict_retriever(Xi,vvXi[j])
                S[i][j]=np.sum(op1*op2)
        G2=0
        N = np.sum(S)
#         print(S)
#         print(N)
        Si =np.sum(S,axis=1)
        Sj =np.sum(S,axis=0)
        Df = ((len(vvTarget)-1)*(len(vvXi)-1))
#         print(S[np.where(S>0)])
        Dedf = len((S[np.where(S>0)]))\
                      - len(Si[np.where(Si>0)])\
                      - len(Sj[np.where(Sj>0)]) +1
#         print("Dedf before if: ",Dedf)
        if (Dedf<1):
            Dedf=1            
            #return 0
        #print(S)
        for i in range(0,len(vvTarget)):
            for j in range(0,len(vvXi)):
                #print(i)
                #print(j)
                if (S[i][j]>0):
                    #print(Si[i])
                    #print(Sj[j])
                    G2 = G2 + S[i][j]*np.log((S[i][j])*N/(Si[i]*Sj[j]))
                    #print(G2)
        
    else: # test conditional dependency
        S = np.empty((len(CondVars)+2,len(vvTarget),len(vvXi),szCondVars,data_c_size))
        for l in range(0,len(CondVars)+2):
            for i in range(0,len(vvTarget)):
                for j in range(0,len(vvXi)):
                    for k in range(0,szCondVars):
                        if (l<len(CondVars)):
                            condValue = GetCondValues(k,CondVars)                        
                            X_l = CondVars[l]
                            op_x = dep_dict_retriever(X_l,condValue[l])
                            S[l][i][j][k]=op_x
                        else:
                            if (l == len(CondVars)):
                                S[l][i][j][k] = dep_dict_retriever(TargetNode,vvTarget[i])
                            else:
                                S[l][i][j][k] = dep_dict_retriever(Xi,vvXi[j])
                        #S[i][j][k]=np.sum(flagDataCondVars)
        # to gpu
        cond_gpu_mdarray = gpuarray.to_gpu(np.asarray(S).astype(np.float32))
        S = (reduce(lambda x,y:x*y, cond_gpu_mdarray)).get()
        S_post = np.empty((len(vvTarget),len(vvXi),szCondVars))
        for i, j, k in itertools.product(range(len(vvTarget)), 
                                         range(len(vvXi)), 
                                         range(szCondVars)):
            S_post[i][j][k]=np.sum(S[i][j][k])
        #print(S_post)
        G2 = 0
        
        Sjk = np.sum(S_post,axis=0)
        Sik = np.sum(S_post,axis=1)
        Sk = np.sum(Sjk,axis=0)
        
        #Sjk = sum(S,1);
        #Sik = sum(S,2);
        #Sk = sum(Sjk,2); 
        Dedf = len(S_post[np.where(S_post>0)]) -\
                len(Sik[np.where(Sik>0)]) -\
                len(Sjk[np.where(Sjk>0)]) +\
                len(Sk[np.where(Sk>0)])
        if Dedf<1:
            Dedf=1
            #print('No, here is the trouble')
            #return 0
        
        # compute G2 statistic
        #print(S.shape)
        #print(Sjk.shape)
        #print(Sik.shape)
        #print(Sk.shape)
        for i in range(0,len(vvTarget)):
            for j in range(0,len(vvXi)):
                for k in range(0,szCondVars):
                    if (S_post[i][j][k]>0):
                        #print(i,j,k)
                        #test_op =  Sk[k]
                        #test_op = Sik[i][k]
                        #test_op = Sjk[j][k]
                        G2 = G2 + (S_post[i][j][k] * np.log(\
                                    (S_post[i][j][k] * Sk[k])/\
                                    (Sik[i][k] * Sjk[j][k])))
        G2 = 2*G2
    
    
    
    assoc = (alpha - (1 - chi2.cdf(G2,Dedf)))/alpha
    #print(G2)
    #print(Dedf)
    #print(assoc)
    if assoc<0:
        assoc=0
#     print("Target Node: ",TargetNode['name'])
#     print("Node XI: ", Xi['name'])
#     print("\tG2:",G2)
#     print("\tDedf: ",Dedf)
#     print("\tAssoc: ",assoc)
#     print()
#     print()
    return assoc
```

```python
def getZsubset(bin_id,Z):
    bin_str=str(bin_id[::-1])
    Zsubset=[]
    for i in range(0,len(bin_str)):
        if bin_str[i]=='1':
            Zsubset.append(Z[i])
    return Zsubset
```

```python
def arrayUniverse(TargetNode,X):
    Universe = []
    for key in X:
        #print("Key in X:",key)
        #print("Target Node is: ",TargetNode)
        if (key!=TargetNode):
            append_dict={'name':key,'data':X[key].copy(deep=True).tolist()}
            Universe.append(append_dict)
    return Universe
```

```python
def MaxMinHeuristic(TargetNode, CPC, Universe, X, alpha):
    F_arr=[]
    Z = CPC
    fixedCondVars = []
    if (len(CPC)>0):
        Z = CPC[0:-1]           # all but the last one   
        fixedCondVars = [CPC[-1]] # we use last one
    for i in range(len(Universe)-1,-1,-1):
        if (len(Universe[i])==0):
            continue
        assoc = MinAssoc(TargetNode,Universe[i],Z,fixedCondVars,X,alpha)
        if (assoc>0):
            F_arr.append([Universe[i],assoc])
        if (assoc==0):
            Universe.pop(i)
            
    return [F_arr,Universe]
```

```python
def MMPC_v1(TargetNode,Universe,X,alpha):
    
    CPC=[]
    #print(TargetNode)
    # Phase I: Foward
    print("Entering Phase I")
    print("MMPC_beggining: \n"+str(len(Universe)))
    while len(Universe)>0:
        CPC_old = list(CPC) # copy
        maxminheur=MaxMinHeuristic(TargetNode,CPC,list(Universe),X,alpha)
        # maxminheur = [F, assocF, Universe]
        F = maxminheur[0]
        assocF = maxminheur[1]
        #print(assocF)
        Universe = maxminheur[2]
        #print("Universe printing line")
        #print(Universe)
        #print("\n=============================")
        #print(CPC)
        if assocF > 0:
            CPC.append(F)
            indF=Universe.index(F)
            Universe.pop(indF)
        #if (len(CPC)==len(CPC_old)) or (len(CPC)>0.3*(len(Universe)-1)):
        if (len(CPC)==len(CPC_old)):
            break
        print("\nUniverse actual size:")
        print(len(Universe))
        print("CPC actual size:")
        print(len(CPC))
        print("CPC contents:")
        print_CPC(CPC)
    
    # Phase 2: Backward
    print("\nEntering Phase II")
    CPC=CPC[::-1]
    if len(CPC)>1:
        Z=list(CPC)
        for i in range(len(CPC)-1,-1,-1):
            # index is i
            Z.pop(i)
            #print("Analyzing D-separator for ",CPC[i]['name'])
            if ExistDseparator(TargetNode,CPC[i],Z,X,alpha)[0] == True:
                #print("it did exist! removing from cpc.")
                CPC.pop(i)
#     return symmetry_test_v2(TargetNode,CPC,X,alpha)
    return CPC
```

```python
def MMPC_(TargetNode,Universe,X,alpha):
    
    CPC=[]
    #print(TargetNode)
    # Phase I: Foward
    print("Entering Phase I")
    print("MMPC_beggining: \n"+str(len(Universe)))
    while len(Universe)>0:
        CPC_old = list(CPC) # copy
        maxminheur=MaxMinHeuristic(TargetNode,CPC,list(Universe),X,alpha)
        # maxminheur = [F, assocF, Universe]
        F_arr = maxminheur[0]
        if (len(F_arr)>0):
            max_assoc = max([x[1] for x in F_arr])
        else:
            max_assoc = 0
        #print(assocF)
        Universe = maxminheur[1]
        #print("Universe printing line")
        #print(Universe)
        #print("\n=============================")
        #print(CPC)
        if max_assoc > 0:
            # append all nodes that fulfill the requirements
            for nn in [x[0] for x in F_arr if x[1]>=max_assoc]:
                CPC.append(nn)
                indF=Universe.index(nn)
                Universe.pop(indF)
        #if (len(CPC)==len(CPC_old)) or (len(CPC)>0.3*(len(Universe)-1)):
        if (len(CPC)==len(CPC_old)):
            break
        print("\nUniverse actual size:")
        print(len(Universe))
        print("CPC size before filtering: "+str(len(CPC)))
        CPC = backward_phase_v4(TargetNode,CPC,X,alpha,5)
        print("CPC size after filtering: "+str(len(CPC)))
        
#         print("CPC actual size:")
#         print(len(CPC))
#         print("CPC contents:")
#         print_CPC(CPC)
        
    # Phase 2: Backward
    #print("\nEntering Phase II")
#     CPC=CPC[::-1]
#     if len(CPC)>1:
#         Z=list(CPC)
#         for i in range(len(CPC)-1,-1,-1):
#             # index is i
#             Z.pop(i)
#             #print("Analyzing D-separator for ",CPC[i]['name'])
#             if ExistDseparator(TargetNode,CPC[i],Z,X,alpha)[0] == True:
#                 #print("it did exist! removing from cpc.")
#                 CPC.pop(i)
#     return symmetry_test_v2(TargetNode,CPC,X,alpha)
    return CPC
```

```python
def MMPC(TargetNode, X, alpha):
    # The universe will be an array of DataFrame Columns
    Universe = arrayUniverse(TargetNode,X)
    print(len(Universe))
    X = arrayX(X)
    print(len(X))
    for column in X:
        if (column['name']==TargetNode):
            TargetNode = column
            break
    CPC = MMPC_(TargetNode,Universe,X,alpha)
    return CPC
```

```python
def getKeysPartition(targetNode, data_keys,n_partitions):
    n_partitions = int(np.log2(len(data_keys))/2)  # log2 ( number of variables ) divided by 2
    partitions = []
    partition_diff = int(int(len(data_keys)/n_partitions)/2)
    for i in range(0,n_partitions):
        if (i==0):
            partitions.append(list(data_keys[-partition_diff::])+list(data_keys[0:int(len(data_keys)/n_partitions)]))
            continue
        if (i<n_partitions-1):
            partitions.append(list(data_keys[-partition_diff+int(len(data_keys)/n_partitions)*i:int(len(data_keys)/n_partitions)*(i+1)]))
            continue
        if (i==n_partitions-1):
            partitions.append(list(data_keys[-partition_diff+int(len(data_keys)/n_partitions)*i::]))
            
    for i in range(0,n_partitions):
        for j in range(len(partitions[i])-1,-1,-1):
            if (partitions[i][j]==targetNode):
                partitions[i].pop(j)
        partitions[i] = partitions[i] + [targetNode]
    return partitions,n_partitions
```

```python
def backward_phase_v2(TargetNode, CPC, X,alpha):
    #print("\nEntering Phase II")
    if (len(CPC)>1):
        # will divide the sets in 
        Z=list(CPC)
        while len(Z)>0:
            # Will test all vs the Target, and select the lowest
            minDep=999
            minDepIndex=-1
            for i in range(0,len(Z)):
                zCopy = list(Z)
                zCopy.pop(i)
                auxDep = Dep(TargetNode,Z[i],zCopy,X,alpha)
                if (minDep>auxDep):
                    minDep = auxDep
                    minDepIndex=i
            # Find index in CPC
            CPCindex=-1
            for i in range(0,len(CPC)):
                if (CPC[i]['name']==Z[minDepIndex]['name']):
                    CPCindex=i
                    break
            
            # Remove from Z
            Z.pop(minDepIndex)
            #print_names(Z)
            #print("Analyzing D-separator for ",CPC[CPCindex]['name'])
            if ExistDseparator(TargetNode,CPC[CPCindex],Z,X,alpha)[0] == True:
                #print("it did exist! removing from cpc.")
                CPC.pop(CPCindex)
    return CPC
```

```python
def backward_phase_v3(TargetNode, CPC, X,alpha,recursive_flag):
    #print("\nEntering Phase II")
    if (len(CPC)>1):
        # will divide the sets in np.log(size)*2
        size_sets = ceil(np.log(len(CPC))*2)
        #print(size_sets)
        n_sets = ceil(len(CPC)/size_sets)
        nodes_sets = []
        CPC_copy = list(CPC)
        for set_counter in range(0,n_sets):
            sub_node_set = []
            while (len(sub_node_set)<size_sets and len(CPC_copy)>0):
                rand_index = random.randint(0,len(CPC_copy))-1
                sub_node_set.append(CPC_copy[rand_index])
                CPC_copy.pop(rand_index)
            nodes_sets.append(sub_node_set)
        
        pre_cpc = []
        for n_set in nodes_sets:
            for node in backward_phase_v2(TargetNode,n_set,X,alpha):
                pre_cpc.append(node)
    else:
        return CPC
    if (recursive_flag>0):
        return backward_phase_v3(TargetNode,pre_cpc,X,alpha,recursive_flag-1)
    else:
        return pre_cpc
```

```python
def backward_phase_v4(TargetNode, CPC, X,alpha,loops_flag):
    #print("\nEntering Phase II")
    votation_arr = []
    if (len(CPC)>1):
        # will divide the sets in np.log(size)*2
        size_sets = ceil(np.log(len(CPC))*2)
        #print(size_sets)
        n_sets = ceil(len(CPC)/size_sets)
        for k in range(0,loops_flag):
            nodes_sets = []
            CPC_copy = list(CPC)
            for set_counter in range(0,n_sets):
                sub_node_set = []
                while (len(sub_node_set)<size_sets and len(CPC_copy)>0):
                    rand_index = random.randint(0,len(CPC_copy))-1
                    sub_node_set.append(CPC_copy[rand_index])
                    CPC_copy.pop(rand_index)
                nodes_sets.append(sub_node_set)
            pre_cpc = []
            for n_set in nodes_sets:
                for node in backward_phase_v2(TargetNode,n_set,X,alpha):
                    pre_cpc.append(node)
            votation_arr.append(pre_cpc)
        # voting
        returnable_cpc = []
        for node in CPC:
            v_counter=0
            for votation in votation_arr:
                if node in votation:
                    v_counter=v_counter+1
            if (v_counter>loops_flag/2):
                returnable_cpc.append(node)
        return returnable_cpc
    else:
        return CPC
```
